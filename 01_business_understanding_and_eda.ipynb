### Target Variable Definition

The target variable for this analysis is **Churn Label**, which indicates whether a customer has discontinued the service.

- "Yes" â†’ Customer churned
- "No" â†’ Customer retained

Preliminary inspection shows that churned customers represent a minority of the dataset, indicating class imbalance. This will be considered during model evaluation, where recall will be prioritized over accuracy.

## Feature Selection for Modeling


To prepare the dataset for predictive modeling, a subset of features was selected based on business relevance and data leakage considerations.

The following types of variables were excluded:
- Unique identifiers (e.g., CustomerID)
- Geographic variables with high cardinality (e.g., City, Zip Code)
- Variables that directly reveal churn outcomes (e.g., Churn Score, Churn Reason)

The remaining features capture customer tenure, service subscriptions, billing behavior, and contract details, which are appropriate inputs for churn prediction.


```python
target = "Churn Label"

drop_cols = [
    "CustomerID",
    "Count",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Lat Long",
    "Latitude",
    "Longitude",
    "Churn Score",
    "Churn Reason",
    "Churn Value"
]

model_df = df.drop(columns=drop_cols)

model_df.head()
```

```python
X = model_df.drop(columns=target)
y = model_df[target]

X.shape, y.shape
```

```python
categorical_features = X.select_dtypes(include="object").columns.tolist()
numerical_features = X.select_dtypes(exclude="object").columns.tolist()

categorical_features, numerical_features
```

```python
X["Total Charges"] = pd.to_numeric(X["Total Charges"], errors="coerce")
```

```python
X.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

```python
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2, 
    stratify=y, 
    random_state=42
)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        solver="lbfgs"
    ))
])
```

```python
log_reg_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred_lr = log_reg_pipeline.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
confusion_matrix(y_test, y_pred_lr)
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        min_samples_split=5,
        min_samples_leaf=2,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])
```

```python
rf_pipeline.fit(X_train, y_train)
```

```python
y_pred_rf = rf_pipeline.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
confusion_matrix(y_test, y_pred_rf)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
y_proba = rf_pipeline.predict_proba(X_test)[:, 1]
```

```python
import numpy as np
import pandas as pd

risk_df = X_test.copy()
risk_df["actual_churn"] = y_test.values
risk_df["churn_probability"] = y_proba

risk_df["risk_segment"] = pd.cut(
    risk_df["churn_probability"],
    bins=[0, 0.3, 0.6, 1.0],
    labels=["Low Risk", "Medium Risk", "High Risk"]
)

risk_df[["churn_probability", "risk_segment"]].head()
```

```python
risk_df["risk_segment"].value_counts(normalize=True) * 100
```

```python
pd.crosstab(
    risk_df["risk_segment"],
    risk_df["actual_churn"],
    normalize="index"
)
```

### Churn Risk Segmentation

Based on predicted churn probabilities from the Random Forest model, customers were segmented into low, medium, and high-risk groups.

- **High Risk:** Customers with a high likelihood of churn who should be prioritized for immediate retention actions such as targeted offers or proactive support.
- **Medium Risk:** Customers who may churn without intervention and could benefit from engagement campaigns or contract incentives.
- **Low Risk:** Stable customers with low churn risk, suitable for standard communication and upsell opportunities.

This segmentation enables targeted retention strategies rather than uniform outreach, improving operational efficiency and customer experience.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=risk_df, x="risk_segment", order=["Low Risk", "Medium Risk", "High Risk"])
plt.title("Customer Distribution by Churn Risk Segment")
plt.xlabel("Risk Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
churn_rate_by_segment = (
    risk_df.groupby("risk_segment")["actual_churn"]
    .apply(lambda x: (x == "Yes").mean())
    .reset_index(name="churn_rate")
)

sns.barplot(data=churn_rate_by_segment, x="risk_segment", y="churn_rate",
            order=["Low Risk", "Medium Risk", "High Risk"])
plt.title("Observed Churn Rate by Risk Segment")
plt.ylabel("Churn Rate")
plt.xlabel("Risk Segment")
plt.show()

churn_rate_by_segment
```

```python
churn_by_segment = (
    pd.crosstab(
        risk_df["risk_segment"],
        risk_df["actual_churn"],
        normalize="index"
    ) * 100
)

churn_by_segment
```

```python
churn_by_segment["Yes"].plot(kind="bar")
plt.title("Actual Churn Rate by Risk Segment")
plt.ylabel("Churn Rate (%)")
plt.xlabel("Risk Segment")
plt.show()
```

The risk segmentation aligns strongly with observed churn behavior. Customers classified as high risk exhibit substantially higher actual churn rates compared to medium and low-risk groups, validating the effectiveness of probability-based segmentation for targeted retention strategies.
